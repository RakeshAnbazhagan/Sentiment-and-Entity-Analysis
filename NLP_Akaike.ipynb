```python
#importing files
import json
with open('tweets.json') as jfile:
  d = json.load(jfile)
```

```python
#dataframe using pandas
import pandas as pd
df = pd.DataFrame.from_dict(d,orient='index')
df.index.name='tweet_id'
df.head()
```

```python
#removing links
import re
def remove_links(text):
  link= re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+',' ', text)
  return link

df['tweet_text_wo_link']=df['tweet_text'].apply(lambda x: remove_links(x))
df.head()
```

```python
import string
string.punctuation
```

```python
#removing punctuations
def remove_punctuation(text):
    no_punct=[words for words in text if words not in string.punctuation]
    words_wo_punct=''.join(no_punct)
    return words_wo_punct
df['tweet_text_wo_punct']=df['tweet_text_wo_link'].apply(lambda x: remove_punctuation(x))
df.head()
```

```python
def remove_emoji(text):
  review=re.sub('[^a-zA-Z]',' ',text)
  return review

df['tweet_text_wo_emoji']=df['tweet_text_wo_punct'].apply(lambda x: remove_emoji(x))
df.head()
```

```python
#importing sciscpacy for figuring out the entities as it is realted to Medical terms
import scispacy
import spacy
nlp = spacy.load("en_ner_bc5cdr_md")
```

```python
def entity_extraction(text):
  doc=nlp(text)
  results=[ent.text for ent in doc.ents]
  return results
```

```python
df['entities']=df['tweet_text_wo_emoji'].apply(lambda x: entity_extraction(x))
```

```python
x=pd.Series(df['entities'])
```

```python
entities = x.value_counts(dropna=True, sort=True).to_frame()
```

```python
entities.index.name='entity'
entities.rename(columns = {'entities':'frequency'}, inplace = True)
```

```python
objective_1=entities.iloc[1:, :]
```

```python
#objective_1
objective_1.to_csv('objective_1.csv') 
```

```python
from nltk.sentiment.util import *
from textblob import TextBlob
```

```python
#sentiment Analysis
df['tweet_text'] = df['tweet_text'].astype('str')
def get_polarity(text):
  return TextBlob(text).sentiment.polarity
df['Polarity_score'] = df['tweet_text'].apply(get_polarity)
```

```python
def getAnalysis(score):
  if score < 0:
    return 'Negative'
  elif score == 0:
    return 'Neutral'
  else:
    return 'Positive'
  
df['overall_polarity'] = df['Polarity_score'].apply(getAnalysis )
```

```python
objective_2 = pd.concat([df['entities'],df['tweet_author'], df['overall_polarity']], axis=1)
objective_2.reset_index(inplace=True, drop=True)
```
